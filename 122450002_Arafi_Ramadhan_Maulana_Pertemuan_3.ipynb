```python
import pandas as pd
file_path = '/content/data.csv'
data = pd.read_csv(file_path)

cleaned_data = data.drop([0, 1, 2]).reset_index(drop=True)
cleaned_data.columns = ['Kecamatan', 'Laki_2016', 'Laki_2017', 'Laki_2018', 'Perempuan_2016', 'Perempuan_2017', 'Perempuan_2018', 'Total_2016', 'Total_2017', 'Total_2018']

for col in cleaned_data.columns[1:]:
    cleaned_data[col] = pd.to_numeric(cleaned_data[col], errors='coerce')

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))

sns.lineplot(x=[2016, 2017, 2018], y=cleaned_data.loc[0, ['Laki_2016', 'Laki_2017', 'Laki_2018']], label="Laki-laki Sawangan", marker="o")
sns.lineplot(x=[2016, 2017, 2018], y=cleaned_data.loc[1, ['Laki_2016', 'Laki_2017', 'Laki_2018']], label="Laki-laki Bojongsari", marker="o")

sns.lineplot(x=[2016, 2017, 2018], y=cleaned_data.loc[0, ['Perempuan_2016', 'Perempuan_2017', 'Perempuan_2018']], label="Perempuan Sawangan", marker="o")
sns.lineplot(x=[2016, 2017, 2018], y=cleaned_data.loc[1, ['Perempuan_2016', 'Perempuan_2017', 'Perempuan_2018']], label="Perempuan Bojongsari", marker="o")

plt.title("Population Trends in Sawangan and Bojongsari (2016-2018)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.xticks([2016, 2017, 2018])
plt.legend(loc="upper left")
plt.tight_layout()

plt.show()
```

**TUJUAN**\
Tujuan Utama dari visualisasi ini adalah untuk melihat tren dalam waktu tiga tahun, dari 2016 - 2018 di kecamatan Sawangan dan Bojongsari. Visualisasi ini memberikan wawasan tentang:/
1. Pertumbuhan penduduk laki-laki dan perempuan secara terpisah
2. perbandingan pertumbuhan dua kecamatan yang ditampilkan secara terpisah
